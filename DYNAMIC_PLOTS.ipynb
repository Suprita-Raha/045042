```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.offline as py
import bubbly
from bubbly.bubbly import bubbleplot

# Import / Read Dataset
df = pd.read_excel(r"/content/Python Dynamic DataSet.xlsx")

# Display Dataset
df.head()
```

```python
# Creating a moving bubble chart over the Months to display Revenue
fig = px.scatter(df, x='Revenue (USD)', y='Rank', animation_frame='Month',
                 color='Region', size='Revenue (USD)',
                 title='Revenue Over Months Rankwise',
                 labels={'Revenue (USD)': 'Revenue'},
                 template='plotly_dark')

fig.update_layout(xaxis=dict(title='Revenue (USD)'), yaxis=dict(title='Rank'))

fig.show()
```

**1. Objectives:**

-To analyze the monthly revenue performance.

-To understand the rank-wise distribution of revenue.

-To identify regional patterns in revenue growth.

**2. Briefs on the Data & the Analysis:**

-The dataset used for this analysis contains information on revenue, rank, and region.

-We utilized the Plotly library to create an animated scatter plot, with quarters as animation frames.

-The x-axis represents Revenue (USD), the y-axis represents Rank, and the color represents different Regions.

**3. Observations:**

*Overall Revenue Over Months*

-The animation reveals the fluctuation in revenue across quarters.

-Asia consistently leads in revenue, except for a notable surge in the UK in Month 6.

-The USA experiences lower recurring revenue throughout the observed period.

*Rank-wise Distribution*

-Rank-wise distribution provides insights into the performance of different ranks across quarters.

-Notable changes in rank positions may be correlated with shifts in revenue.

*Regional Patterns*

-Asia consistently dominates in revenue, showcasing a robust and stable p-erformance.

The UK sees a substantial increase in revenue in , potentially indicating a specific event or strategic initiative.

-The USA lags in recurring revenue, suggesting potential areas for improvement or targeted strategies.

**4. Managerial Implications or Insights:**

-Asia's Revenue Dominance: The consistent leadership of Asia in revenue indicates a strong market presence. Management should continue to invest in strategies that contribute to this stability.

-UK's Month 6 Surge: The significant increase in the UK's revenue in Month 6 warrants further investigation. Identifying the factors contributing to this surge can help replicate success in other periods.

-USA's Low Recurring Revenue: The lower recurring revenue in the USA signals the need for targeted interventions. Management should explore strategies to enhance revenue streams in this region.
